```python
#_______________TASK: 1________________

a= int(input("Enter 1st number: "))
b= int(input("Enter 2nd number: "))
c= input("Enter Operation: ")
if c == "add":
    print("Answer is: ",a+b)
if c == "subtract":
    print("Answer is: ",a-b)
if c == "multiply":
    print("Answer is: ",a*b)
if c == "divide":
    print("Answer is: ",a/b)
if c == "power":
    print("Answer is: ",a**b)
```

```python
#_________________TASK: 2______________

list = [1, 2, 3, 4, 'a', 'b', 'x', 5, 'z'] 
print ("initial list", str(list)) 
for item in list: 
    try: 
        res= int(item) 
    except ValueError: 
        pass 
print ("Total numeric values in a list are= ", res) 
```
