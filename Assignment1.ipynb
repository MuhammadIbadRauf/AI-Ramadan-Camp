```python
First=int(input("Enter Your 1st Subject Marks: "))
Second=int(input("Enter Your 2nd Subject Marks: "))
Third=int(input("Enter Your 3rd Subject Marks: "))
Fourth=int(input("Enter Your 4th Subject Marks: "))
Fifth=int(input("Enter Your 5th Subject Marks: "))
a=First+Second+Third+Fourth+Fifth
Total =  (a/ 500)*100
if Total >= 80:
   print("YOur Grade is A+ with total percentage of: ", Total)
if Total >=70 and Total < 80:
   print("YOur Grade is A with total percentage of: ", Total)
if Total >= 60 and Total < 70:
   print("YOur Grade is B with total percentage of: ", Total)
if Total >= 50 and Total < 60:
   print("YOur Grade is C with total percentage of: ", Total)
if Total >= 40 and Total < 50:
   print("YOur Grade is D with total percentage of: ", Total)
if Total < 40:
   print("No Grade! Work Hard! Your Percentage is: ", Total)
```
